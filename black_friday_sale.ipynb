```python
#alpha tuning
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
import numpy as np
ridge_regression=Ridge()
alpha_range=np.logspace(-4,4,100)
parameters={'alpha': alpha_range }
regression_scoring=['neg_mean_absolute_error','neg_mean_squared_error','neg_root_mean_squared_error','r2']
for i in regression_scoring:
  ridge_cv=GridSearchCV(ridge_regression,parameters,scoring=i,cv=5)
  ridge_cv.fit(x_train,y_train)
  print(f"Best lambda value for {i} : ,{ridge_cv.best_params_}")
  print(f"Best score cost function for {i} : ,{ridge_cv.best_score_}")
```
